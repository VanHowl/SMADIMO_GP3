```python
import pandas as pd
import numpy as np
```

```python
rests = pd.read_csv('top_75_restaurants.csv')
review_w = pd.read_csv('reviews_with_weather.csv')
holidays = pd.read_csv('us_holidays_2005-2025.csv')
```

```python
merged_df = pd.merge(rests, review_w, left_on='business_id', right_on='business_id')
merged_df.head()
```

```python
columns_to_drop = []
for col in merged_df.columns:
  if col.endswith('_x'):
    original_col_name = col[:-2]
    if original_col_name + '_y' in merged_df.columns:
      columns_to_drop.append(col)

merged_df = merged_df.drop(columns=columns_to_drop)
merged_df.columns = merged_df.columns.str.replace('_y', '') ##str из Индекса в строку для метода
merged_df.head()
```

```python
attributes_series = merged_df['attributes']
def safe_literal_eval(attr_string):
    try:
        return eval(attr_string) ## преобразуем строчку в словарь
    except (SyntaxError, ValueError):
        return {}
attributes_series = attributes_series.apply(safe_literal_eval)
unique_attributes = set()
for attr_dict in attributes_series:
    unique_attributes.update(attr_dict.keys())
unique_attributes_list = list(unique_attributes)
unique_attributes_list
```

```python
for attribute in unique_attributes_list:
    merged_df[attribute] = attributes_series.apply(lambda x: x.get(attribute, None)) ## пробегаемся по всем строчкам
merged_df.head()
```

```python
vocab_attr_columns = []
## ghjdthrf yf gjntywbfkmysq ckjdfhm
for col in merged_df.columns:
    if merged_df[col].apply(lambda x: isinstance(x, str) and ('{' in x or '[' in x)).any():
        try:
            first_non_null = merged_df[col].dropna().iloc[0] ## Проверяем первое не пустое значение
            evaluated_value = eval(first_non_null)
            if isinstance(evaluated_value, (dict, list)):
                vocab_attr_columns.append(col)
        except:
            pass
vocab_attr_columns = [col for col in vocab_attr_columns]
vocab_attr_columns
```

```python
ambience_values = set()
for index, row in merged_df.iterrows(): ##просматриваем все строчки
    ambience_dict_str = row['Ambience']
    if ambience_dict_str and isinstance(ambience_dict_str, str):
        try:
            ambience_dict = eval(ambience_dict_str)
            if isinstance(ambience_dict, dict):
                ambience_values.update(ambience_dict.keys())
        except (SyntaxError, ValueError):
            pass
unique_ambience_list = list(ambience_values)
unique_ambience_list
```

```python
for ambience_type in unique_ambience_list:
    merged_df[ambience_type] = merged_df['Ambience'].apply(
        lambda x: eval(x).get(ambience_type, False) if isinstance(x, str) else False ##создаем столбуц и тут же пробегаемся для него по словарям
    )
merged_df.head()
```

```python
business_parking_values = set()
for index, row in merged_df.iterrows():
    business_parking_dict_str = row['BusinessParking']
    if business_parking_dict_str and isinstance(business_parking_dict_str, str):
        try:
            business_parking_dict = eval(business_parking_dict_str)
            if isinstance(business_parking_dict, dict):
                business_parking_values.update(business_parking_dict.keys())
        except (SyntaxError, ValueError):
            pass
unique_business_parking_list = list(business_parking_values)
## создаем и заполняем значения
unique_business_parking_list
for parking_type in unique_business_parking_list:
    merged_df[parking_type] = merged_df['BusinessParking'].apply(
        lambda x: eval(x).get(parking_type, False) if isinstance(x, str) else False
    )
merged_df.head()
```

```python
hours_values = set()
for index, row in merged_df.iterrows():
    hours_dict_str = row['hours']
    if hours_dict_str and isinstance(hours_dict_str, str):
        try:
            hours_dict = eval(hours_dict_str)
            if isinstance(hours_dict, dict):
                hours_values.update(hours_dict.keys())
        except (SyntaxError, ValueError):
            pass
unique_hours_list = list(hours_values)
unique_hours_list
for day_of_week in unique_hours_list:
    merged_df[day_of_week] = merged_df['hours'].apply(
        lambda x: eval(x).get(day_of_week, None) if isinstance(x, str) else None
    )
merged_df.head()
```

```python
merged_df['categories_list'] = merged_df['categories'].str.split(',')
all_categories = merged_df['categories_list'].explode().unique()
category_one_hot = pd.get_dummies(merged_df['categories_list'].explode()).groupby(level=0).sum().astype(bool) ##explode для возвращения серии далее собираем one-hot encoding обратно по исходным строкам по исходным индексам (т.к. создается слишком много строк)
merged_df = pd.concat([merged_df, category_one_hot], axis=1)
display(merged_df.head())
```

```python
days_of_week = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
for day in days_of_week:
    if day in merged_df.columns:
        merged_df[[f'{day}_open', f'{day}_close']] = merged_df[day].str.split('-', expand=True) ##expand=True сразу +2 столбца
        merged_df = merged_df.drop(columns=[day])
merged_df.head()
```

```python
for day in days_of_week:
    if f'{day}_open' in merged_df.columns:
        merged_df[f'{day}_open'] = pd.to_datetime(merged_df[f'{day}_open'], format='%H:%M', errors='coerce').dt.time # пропуски заполнин NaN
    if f'{day}_close' in merged_df.columns:
        merged_df[f'{day}_close'] = pd.to_datetime(merged_df[f'{day}_close'], format='%H:%M', errors='coerce').dt.time
merged_df.head()
```

```python
best_nights_values = set()
for index, row in merged_df.iterrows():
    best_nights_dict_str = row['BestNights']
    if best_nights_dict_str and isinstance(best_nights_dict_str, str):
        try:
            best_nights_dict = eval(best_nights_dict_str)
            if isinstance(best_nights_dict, dict):
                best_nights_values.update(best_nights_dict.keys())
        except (SyntaxError, ValueError):
            pass
unique_best_nights_list = list(best_nights_values)
unique_best_nights_list
for night_type in unique_best_nights_list:
    merged_df[night_type] = merged_df['BestNights'].apply(
        lambda x: eval(x).get(night_type, False) if isinstance(x, str) else False
    )
merged_df.head()
```

```python
music_values = set()
for index, row in merged_df.iterrows():
    music_dict_str = row['Music']
    if music_dict_str and isinstance(music_dict_str, str):
        try:
            music_dict = eval(music_dict_str)
            if isinstance(music_dict, dict):
                music_values.update(music_dict.keys())
        except (SyntaxError, ValueError):
            pass
unique_music_list = list(music_values)
unique_music_list

for music_type in unique_music_list:
    merged_df[music_type] = merged_df['Music'].apply(
        lambda x: eval(x).get(music_type, False) if isinstance(x, str) else False
    )
merged_df.head()
```

```python
goodformeal_values = set()
for index, row in merged_df.iterrows():
    goodformeal_dict_str = row['GoodForMeal']
    if goodformeal_dict_str and isinstance(goodformeal_dict_str, str):
        try:
            goodformeal_dict = eval(goodformeal_dict_str)
            if isinstance(goodformeal_dict, dict):
                goodformeal_values.update(goodformeal_dict.keys())
        except (SyntaxError, ValueError):
            pass
unique_goodformeal_list = list(goodformeal_values)
unique_goodformeal_list
for meal_type in unique_goodformeal_list:
    merged_df[meal_type] = merged_df['GoodForMeal'].apply(
        lambda x: eval(x).get(meal_type, False) if isinstance(x, str) else False
    )
merged_df.head()
```

```python
merged_df = merged_df.drop(columns=['attributes', 'hours', 'categories', 'Ambience', 'BusinessParking', 'categories_list', 'Music', 'BestNights', 'GoodForMeal'])
merged_df.head()
```

```python
unique_states = set()
for index, row in holidays.iterrows():
    counties_str = row['counties']
    if isinstance(counties_str, str):
        states = counties_str.split(';')
        for state in states:
            cleaned_state = state.strip().replace('US-', '')
            unique_states.add(cleaned_state)
unique_states_list = list(unique_states)
unique_states_list
```

```python

for state in unique_states_list:
    holidays[state] = False
for index, row in holidays.iterrows():
    if row['global']:
        for state in unique_states_list:
            holidays.at[index, state] = True
    else:
        counties_str = row['counties']
        if isinstance(counties_str, str):
            states_in_row = [state.strip().replace('US-', '') for state in counties_str.split(';')]
            for state in unique_states_list:
                if state in states_in_row:
                    holidays.at[index, state] = True
holidays.head()
```

```python
import pandas as pd
merged_df['date'] = pd.to_datetime(merged_df['date']).dt.date
holidays['date'] = pd.to_datetime(holidays['date']).dt.date
def is_holiday(row):
    date = row['date']
    state = row['state']
    holiday_row = holidays[holidays['date'] == date]
    if not holiday_row.empty and state in holiday_row.columns:
        return holiday_row[state].iloc[0]
    return False
merged_df['is_holiday'] = merged_df.apply(is_holiday, axis=1)
merged_df.head()
```

```python
weather_mapping = {
    0: 'clear',
    1: 'sunny',
    2: 'overcast',
    3: 'cloudy',
    51: 'light_rain',
    53: 'moderate_rain',
    55: 'heavy_rain',
    61: 'light_shower',
    63: 'rain',
    65: 'heavy_shower',
    71: 'light_snow',
    73: 'snow',
    75: 'heavy_snow'
}

def map_weather_code(code):
    if pd.isna(code):
        return 'неизвестно'
    code = int(code)
    return weather_mapping.get(code, 'неизвестно')

merged_df['weather_description'] = merged_df['weather_code'].apply(map_weather_code)

print(merged_df['weather_description'].value_counts())


weather_dummies = pd.get_dummies(merged_df['weather_description'], prefix='weather')

merged_df = merged_df.drop(['weather_code','weather_description'], axis=1, errors='ignore')

merged_df = pd.concat([merged_df, weather_dummies], axis=1)

print("One-hot-encoded погодные признаки:")
print(weather_dummies.columns.tolist())
print("\nПервые 5 строк с новыми признаками:")
print(merged_df[weather_dummies.columns].head())
```
